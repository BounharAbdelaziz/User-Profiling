```python
data_frame = pd.read_csv('Sentiment.csv', encoding='ISO-8859-1', error_bad_lines=False, delimiter =",")
#data_frame = pd.read_csv("training.txt",delimiter="\t", header=None)
```

```python
data_frame.columns = ['MessageID', 'Sentiment', 'Message' ]
#data_frame.columns = ['Sentiment', 'Message' ]
```

# Scoring the dataset

```python
from afinn import Afinn
afinn = Afinn()
```

```python
def quality(score):
    if score <= -3 :
        return "very disliked"
    if score == -2 :
        return "sad"
    if score == -1 :
        return "dislike"
    if score == 0 :
        return "neutral"
    if score == 1 :
        return "tendency"
    if score == 2 :
        return "likely"
    if score == 3 :
        return "very likely"
    if score >= 4 :
        return "very happy"
    
```

## Example on one row

```python
message = data_frame['Message'][0]
print("[INFO] message : ",message)
afinn.score(message)
```

## Running the scoring through all the dataset 

```python
taille = 500
for i in range(0,taille):
    message = data_frame['Message'][i]
    score = afinn.score(message)
    q = quality(score)
    data_frame.loc[i , 'Sentiment'] = q
    print("[INFO] FOR THE MESSAGE :",message," ***** QUALITY = ",q)
```
